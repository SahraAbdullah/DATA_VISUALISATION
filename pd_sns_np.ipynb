# Returns any null values.

```python
df = pd.DataFrame(titanic)
null_values = df.isnull()
null_counts = null_values.sum()
df['null_count'] = null_counts
print(df)
```

```python
cleaned = df.dropna(subset=['Ticket'])
print(cleaned)
```
