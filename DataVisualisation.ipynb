Data Visualisation of Titanic Dataset

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset('titanic')
```

```python
for column in ['pclass', 'age', 'fare']:
    plt.style.use('_mpl-gallery')
    x = 'fare'
    y = 'pclass', 'age'
    fig, ax = plt.subplots()
    plt.show()
    
```

I used the count plot to count the survivors. I found that using the hue parameter can split the data by a categorical variable. e.g. hue='sex' will return male or female. 

```python
sns.countplot(x='survived', hue='sex', data=titanic)
plt.title('Count of survivors')
plt.show()
```

```python
pie(x='age', data=titanic)
plt.title('Something')
plt.show()
```

```python
sns.violinplot(x="age", y="fare", data=titanic)
plt.title('Fares and Ages')
plt.show()
```

```python
sns.color_palette(palette='PRGn')
sns.stripplot(x="class", y="fare", data=titanic)
plt.title('Fares and Ages')
plt.show()
```

```python
sns.violinplot(x='class', y='age', data=titanic)
plt.title('Distribution of age')
plt.show()
```
